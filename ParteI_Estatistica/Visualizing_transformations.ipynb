# **Visualizing Linear and Non-Linear Transformations**

```
import matplotlib.pyplot as plt
import numpy as np
import random
#
# Random 2D gaussian distribution in a plane
#

plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
npoints=1000
c=np.arange(npoints)
x = np.random.normal(size=[npoints, 2])
plt.scatter(x[:,0],x[:,1], label='Normal Distribution', c=c, cmap = 'plasma')
plt.show()
plt.close()
#plt.savefig("name")
print("X=",x)
print(" ")
print(" ")


```

```
#
# Linear Transformation of the distribution acording to T ( stretching along X and compressing along Y)
#
s1=1000
s4=0.00001
T=np.array([[s1,0],[0,s4]])
print("Matrix T=",T)
print(" ")
print(" ")

y=np.tensordot(x,T,axes=1)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
x = np.random.normal(size=[npoints, 2])
plt.scatter(y[:,0],y[:,1], c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

```
#
# Linear Transformation of the distribution acording to T ( compressing along X and stretching along Y)
#
s1=0.00001
s4=1000.
T=np.array([[s1,0],[0,s4]])
print("Matrix T=",T)
print(" ")
print(" ")

y=np.tensordot(x,T,axes=1)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

```
#
# Linear Transformation of the distribution acording to T ( compressing along X and stretching along Y and rotating )
#
s1=0.00001
s2=0
s3=-1
s4=1000.
T=np.array([[s1,s2],[s3,s4]])
print("Matrix T=",T)
print(" ")
print(" ")

y=np.tensordot(x,T,axes=1)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

```
#
# Non-Linear transform the distribution ( TANH stretches the gaussian distribution to a square region )
#
y=np.tanh(x)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```

```
#
# Non-Linear transform the distribution ( TANH stretches distribution to a square and zom factor squezes to the edges )
#
zoom=3.5
y=np.tanh(zoom*x)
print("Y=",y)
plt.figure(figsize=(5,5),dpi=100)
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(y[:,0],y[:,1], c=c, cmap = 'plasma', label='Transformed Normal Distribution')
plt.show()
plt.close()
#
```
